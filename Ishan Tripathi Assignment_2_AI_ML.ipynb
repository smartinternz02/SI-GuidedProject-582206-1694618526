###1- Download and load the dataset

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/House Price India.csv')
df.head()
```

###2-Perform Univariate Analysis

```python
sns.displot(df.lot_area_renov)
```

```python
sns.barplot(x=df.Date.value_counts().index,y=df.Date.value_counts())
```

```python
palette = sns.color_palette("pastel", len(df['number of bedrooms'].value_counts()))
plt.pie(df['number of bedrooms'].value_counts(),autopct='%1.1f%%',shadow=True, colors=palette)
plt.title('Number of Bedrooms')
plt.show()
```

###3-Perform Bivariate Analysis

```python
sns.lineplot(y=df['Area of the house(excluding basement)'], x=df['number of bedrooms'])
```

```python
sns.scatterplot(x=df['number of floors'],y=df['Price'])
```
